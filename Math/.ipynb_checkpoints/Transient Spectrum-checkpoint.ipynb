```python
import pandas as pd  
import numpy as np  
from scipy.ndimage import gaussian_filter1d  
import matplotlib.pyplot as plt  
import os  

# 读取Excel文件  
file_path = r"D:\Work\Transient Spectrum\TCTA瞬态EL.xls"  # 替换为你的Excel文件路径  
df = pd.read_excel(file_path)  
  
# 定义光滑函数  
def smooth_data(data, sigma=80):  
    return gaussian_filter1d(data, sigma)  
  
# 提取数据  
x = df.iloc[:, 0].values  
y_raw = df.iloc[:, 1:8].values  
y_normalized = df.iloc[:, 8:15].values  
  
# 平滑数据  
y_raw_smoothed = np.array([smooth_data(y) for y in y_raw.T]).T  
y_normalized_smoothed = np.array([smooth_data(y) for y in y_normalized.T]).T   

# 找到每个数组的 NaN 值的位置
nan_indices_raw = np.where(np.isnan(y_raw_smoothed))[0]
nan_indices_norm = np.where(np.isnan(y_normalized_smoothed))[0]

# 在每个 NaN 值的位置上插入一个数据点
for nan_index in nan_indices_raw:
    y_raw_smoothed_padded = np.insert(y_raw_smoothed, nan_index, y_raw_smoothed[nan_index])

for nan_index in nan_indices_norm:
    y_normalized_smoothed_padded = np.insert(y_normalized_smoothed, nan_index, y_normalized_smoothed[nan_index])

# 将平滑后的数据写回新的Excel文件  
smoothed_df = df.copy()  
smoothed_df.iloc[:, 1:8] = y_raw_smoothed  
smoothed_df.iloc[:, 8:15] = y_normalized_smoothed  
smoothed_file_path = r"D:\PythonProject\Math\smoothed_data\smoothed_data(TACA).xlsx"  # 替换为你想保存的文件路径  
smoothed_df.to_excel(smoothed_file_path, index=False)  
  
# 创建保存图像的文件夹（如果不存在）  
images_folder = 'smoothed_images(TACA)'  
os.makedirs(images_folder, exist_ok=True)  

# 计算RMSE
def calculate_rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

# 存储RMSE结果
rmse_results = []
  
for i in range(7):  
    rmse = calculate_rmse(y_raw[:, i], y_raw_smoothed[:, i])
    rmse_results.append(rmse)
    print(f'RMSE for column {i+2}: {rmse}')
    
    plt.figure(figsize=(10, 5))  
    plt.plot(x, y_raw[:, i], label='Raw Data', alpha=0.5)  
    plt.plot(x, y_raw_smoothed[:, i], label='Smoothed Data', linewidth=2)  
    plt.title(f'Smoothed Data (Column {i+2}, RMSE: {rmse:.2f})')  
    plt.xlabel('X Axis')  
    plt.ylabel('Y Axis')  
    plt.legend()  
    image_path = os.path.join(images_folder, f'smoothed_data_col_{i+2}.jpg')  
    plt.savefig(image_path, dpi=300)
    plt.close()  
  
    rmse = calculate_rmse(y_normalized[:, i], y_normalized_smoothed[:, i])
    rmse_results.append(rmse)
    print(f'RMSE for column {i+9}: {rmse}')
    
    plt.figure(figsize=(10, 5)) 
    plt.plot(x, y_normalized[:, i], label='Raw Normalized Data', alpha=0.5)  
    plt.plot(x, y_normalized_smoothed[:, i], label='Smoothed Normalized Data', linewidth=2)  
    plt.title(f'Smoothed Normalized Data (Column {i+9}, RMSE: {rmse:.2f})')  
    plt.xlabel('X Axis')  
    plt.ylabel('Normalized Y Axis')  
    plt.legend()  
    image_path = os.path.join(images_folder, f'smoothed_normalized_data_col_{i+9}.jpg')  
    plt.savefig(image_path, dpi=500)  
    plt.close()  

print("Smoothed data and images have been saved successfully.")
```
